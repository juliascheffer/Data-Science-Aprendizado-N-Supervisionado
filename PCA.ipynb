```python
list_drop_columns = ['grau',
                     'turno',
                     'curso_busca',
                     'curso_id',
                     'bolsa_parcial_ampla',
                     'bolsa_parcial_cotas',
                     'bolsa_integral_ampla',
                     'bolsa_integral_cotas',
                     'mensalidade',
                     'cidade_busca',
                     'uf_busca',
                    'universidade_nome',
                     ]

df_aux = df.drop(columns=list_drop_columns)
df_aux.info()
```

```python
df_aux.head()
```

```python
# Label Encoder

from sklearn.preprocessing import LabelEncoder
```

```python
df_aux.head()
```

```python
df_pca = df_aux.iloc[:,:].values
df_pca
```

```python
le = LabelEncoder()
df_pca = le.fit_transform(df_pca[:,1])
df_pca
```

```python
list(le.classes_)
```

```python
df_aux['nome'] = df_pca
df_aux.head()
```

```python
inverted = le.inverse_transform(df_pca)
print(inverted)
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=2)
```

```python
pca_transform = pca.fit_transform(df_aux)
```

```python
pca_transform.shape
```

```python
# Componentes principais
pca_transform
```

```python
# razão das variáveis explicativas
pca.explained_variance_ratio_
```

```python
# Soma das variáveis explicativas
pca.explained_variance_ratio_.sum()
```

```python
import plotly.express as px
```

```python
labels = {
    str(i): f"PC {i+1} ({var:.1f}%)"
    for i, var in enumerate(pca.explained_variance_ratio_ * 100)
}

fig = px.scatter_matrix(
    pca_transform,
    labels=labels,
    dimensions=range(2),
    color=df_aux["nome"]
)
fig.update_traces(diagonal_visible=False)
fig.show()
```
